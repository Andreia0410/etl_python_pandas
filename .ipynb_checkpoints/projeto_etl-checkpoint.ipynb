# Projeto de ETL

### Proposta do projeto: Armazenar dados de gastos(até o momento), por deputado federal, desde a primeira camada da esteira até a última camada de armazenamento, importando dados da api da camera dos deputados.

#### 1. Obter os dados via api
#### 2. Fazer a limpeza e transformação dos dados
#### 3. Armazenar os dados no formato de DataFrame no Sqlite

### Bibliotecas, frameworks e tecnologias utilizadas

#### 1. Biblioteca pandas;
#### 2. Banco Sqlite;
#### 3. Biblioteca requests

```python
import pandas as pd
import requests
```

```python
url = "https://dadosabertos.camara.leg.br/api/v2/deputados?ordem=ASC&ordenarPor=nome"
resultado = requests.get(url)
```

```python
deputados = pd.DataFrame(resultado.json()['dados'])
deputados.head(10)
```

```python
# Verificando a quantidade de linhas e colunas no DataFrame
deputados.shape
```

```python
gastos = []
for id in deputados.id:
    url_despesa = "https://dadosabertos.camara.leg.br/api/v2/"
    url_despesa_completa = url_despesa + "deputados/" + str(id) + "/despesas?ordem=ASC&ano=2022"
    resposta = requests.get(url_despesa_completa)
    gasto = pd.DataFrame(resposta.json()['dados'])
    gasto['id'] = id
    gastos.append(gasto)
    
```

```python
len(gastos)
```

```python
gastos[0]
```

# Transformação

```python
total = pd.concat(gastos)
```

```python
total.sample(3)
```

# Carregamento de Dados no SQLite

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///dados_abertos.db')
```

```python
deputados.to_sql('deputados_db', engine)
```

```python
total.to_sql('depesas_total', engine)
```

## Análise dos dados carregados

```python
total = pd.read_sql('select * from despesas_total', engine)
```

```python
total.head()
```

```python
# Pegar o nome dos deputados, recuperando a lista de deputados e fazendo um merge

deputados= pd.read_sql('select * from deputados_db', engine)
```

```python
deputados.head()
```

```python
total = total.merge(deputados, on=['id'])
```

```python
total
```

```python
# Verificando os gastos dos deputados
gastos_deputados = total.groupby(['nome']).sum()[['valorDocumento', 'valorLiquido']]
```

```python
# Agrupando por valorDocumento e valorLiquido, os 10 deputados que mais gastaram
gastos_deputados = gastos_deputados.sort_values('valorDocumento', ascending=False)
gastos_deputados.head(10)
```

```python
# Agrupando por valorDocumento e valorLiquido, os 10 deputados que menos gastaram
gastos_deputados.tail(10)
```
